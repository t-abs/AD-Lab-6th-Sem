5.Replace Null values with mean and median

```python
actual_column_name = 'Price'
df[actual_column_name] = df[actual_column_name].fillna(df[actual_column_name].mean())
print(df)

```

```python
actual_col_name='Price'
df[actual_col_name]=df[actual_col_name].fillna(df[actual_col_name].median())
print(df)
```

```python
import pandas as pd

def convert_duration_to_minutes(duration):
  if isinstance(duration, str):
    if 'h' in duration and 'm' in duration:
      hours, minutes = duration.split('h')
      minutes = minutes.replace('m', '')
      total_minutes = int(hours) * 60 + int(minutes)
    elif 'h' in duration:
      total_minutes = int(duration.replace('h', '')) * 60
    elif 'm' in duration:
      total_minutes = int(duration.replace('m', ''))
    else:
      return float('nan')
    return total_minutes
  else:
    return duration

df['Duration'] = df['Duration'].apply(convert_duration_to_minutes)

grouped_data = df.groupby('Airline').agg({
    'Price': ['mean', 'median', 'min', 'max'],
    'Duration': ['mean', 'median']
})

# Display the results
print(grouped_data)
```
